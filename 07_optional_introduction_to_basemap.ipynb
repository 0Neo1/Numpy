```python
maps=Basemap()                                                                  # load map (world)
maps.drawcoastlines(linewidth=0.5,color='black',linestyle='-.')                 # draw boundaries
plt.show()                                                                      # make show plot
```

```python
maps=Basemap()                                                                  # load map (world)
maps.drawmapboundary(fill_color='skyblue')                                      # fill color in map (background)
maps.fillcontinents(color='green',lake_color='blue')                            # fill color in continents
maps.drawcoastlines(linewidth=0.5,color='black',linestyle='solid')              # draw boundaries
# make draw parallels (-90 to 90 degree) and meridians (-180 to 180 degree) with 10 degree gaps
maps.drawparallels(np.arange(-90,90,10))                                        # draw and label parallels (latitude lines)
maps.drawmeridians(np.arange(-180,180,30))                                      # draw and label meridians (longitude lines)
plt.show()                                                                      # make show plot
```

##### [Gnomonic Projection](https://matplotlib.org/basemap/users/gnomon.html)

```python
map1=Basemap(width=15.e6,height=15.e6,projection='gnom',lat_0=77.229446,lon_0=28.612894)# load map for given values
map1.drawmapboundary(fill_color='aqua')                                         # fill background
map1.drawcoastlines(linewidth=1.0,linestyle='solid',color='k')                  # draw coasts lines
map1.fillcontinents(color='coral',lake_color='skyblue')                         # fill continents
map1.drawparallels(np.arange(10,90,20))                                         # draw and label parallels (latitude lines)
map1.drawmeridians(np.arange(-180,180,30))                                      # draw and label meridians (longitude lines)
plt.title('Gnomonic Projection')                                                # add title to plot
plt.show()                                                                      # make plot show
```

##### [Orthographic Projection](https://matplotlib.org/basemap/users/ortho.html)

```python
# lon_0, lat_0 are the center point of the projection.
# resolution = 'l' means use low resolution coastlines. resolution: The resolution of the included coastlines, lakes, and so on.
#   The options are c (crude, the default), l (low), i (intermediate), h (high), f (full) or None.
# NOTE: resolution = 'h' -> OSError: Unable to open boundary dataset file. Only the 'crude', 'low' and 'intermediate' resolution
#   datasets are installed by default. If you are requesting a 'high' or 'full' resolution dataset, you need to install the
#   `basemap-data-hires` package with `conda install -c conda-forge basemap-data-hires`.
map1=Basemap(projection='ortho',lat_0=77.229446,lon_0=28.612894,resolution='i') # load map for given values
map1.drawcoastlines(linewidth=1.0,linestyle='solid',color='g')                  # draw coasts lines
map1.fillcontinents(color='#70543e',lake_color='skyblue')                       # fill continents
map1.drawparallels(np.arange(-90.,120.,30.))                                    # draw and label parallels (latitude lines)
map1.drawmeridians(np.arange(0.,420.,60.))                                      # draw and label meridians (longitude lines)
map1.drawmapboundary(fill_color='aqua')                                         # draws the earth boundary on the map, with optional filling
plt.title('Full Disk Orthographic Projection')                                  # # add title to plot
plt.show()                                                                      # make plot show
```

#### basic examples - plotting by [Passing the bounding box](https://basemaptutorial.readthedocs.io/en/latest/basemap.html#passing-the-bounding-box) geographical longitude and latitude bounding boxes

The way to set the extent is by defining the map bounding box in geographical coordinates:

Argument Description:
* llcrnrlon:	The lower left corner geographical longitude
* llcrnrlat:	The lower left corner geographical latitude
* urcrnrlon:	The upper right corner geographical longitude
* urcrnrlat:	The upper right corner geographical latitude

```python
# Coordinates of India (source): https://en.wikipedia.org/wiki/Geography_of_India
maps=Basemap(
    projection='merc',                                                          # any projection type
    llcrnrlat=5,                                                                # lower-left corner latitude
    urcrnrlat=38,                                                               # upper-right corner latitude
    llcrnrlon=65,                                                               # lower-left corner longitude
    urcrnrlon=100,                                                              # upper-right corner longitude
    resolution='l'                                                              # 'l' for low resolution
)

maps.drawmapboundary(fill_color='aqua')                                         # fill map boundary (background)
maps.drawcoastlines()                                                           # make draw coastlines
maps.drawcountries()                                                            # make draw countries
maps.drawstates()                                                               # make draw states
# make draw parallels (0 to 45 degree) and meridians (60 to 110 degree) with 4 degree gaps
maps.drawparallels(np.arange(0,45,4))                                           # draw and label parallels (latitude lines)
maps.drawmeridians(np.arange(60,110,4))                                         # draw and label meridians (longitude lines)
maps.fillcontinents(color='lightgray',lake_color='lightblue')                   # fill the land with a color (optional)
plt.title('Map of India')                                                       # add title to map
plt.show()                                                                      # make show map
```

#### basic examples - plotting by [Passing the bounding box](https://basemaptutorial.readthedocs.io/en/latest/basemap.html#passing-the-bounding-box) by latitude and longitude

The bounding box giving the center point in geographical coordinates, and the width and height of the domain in the projection units.

Argument Description:
* width:	The width of the map in the projection units
* height: The height of the map in the projection units
* lon_0:	The longitude of the center of the map
* lat_0:	The latitude of the center of the map

```python
longitude:float=78.8718                                                         # longitude of the center of the map (India)
latitude:float=21.7679                                                          # latitude of the center of the map (India)
height:int=4_000_000                                                            # height of map, must be in meters (3214 km height)
width:int=3_000_000                                                             # width of map, must be in meters (2933 km width)
```

```python
maps=Basemap(width=width,height=height,projection='lcc',lat_0=latitude,lon_0=longitude)# load map for given values
plt.figure(figsize=(10,8))                                                      # adjust the size as needed
maps.drawmapboundary(fill_color='aqua')                                         # fill background
maps.drawcoastlines(linewidth=0.5)                                              # draw coasts
maps.fillcontinents(color='green',lake_color='skyblue')                         # fill continents
maps.drawparallels(np.arange(0,45,4))                                           # draw and label parallels (latitude lines) for values (in degrees) given in the sequence circles
maps.drawmeridians(np.arange(60,110,4))                                         # draw and label meridians (longitude lines) for values (in degrees) given in the sequence meridians
x,y=maps(longitude,latitude,inverse=False)                                      # get x and y coordinates for given longitude and latitude
plt.plot(x,y,'r*')                                                              # make plot center dot
plt.title('Lambert Conformal Conic Projection')                                 # add title to plot
plt.show()                                                                      # make plot show
```

```python
longitude:float=78.8718                                                         # longitude of the center of the map
latitude:float=21.7679                                                          # latitude of the center of the map
map1=Basemap(projection='aeqd',lat_0=longitude,lon_0=latitude)                  # load map for given values
map1.drawmapboundary(fill_color='aqua')                                         # fill background
map1.drawcoastlines(linewidth=0.5)                                              # draw coasts lines
map1.fillcontinents(color='green',lake_color='skyblue')                         # fill continents
map1.drawparallels(np.arange(-80,81,20),color='k',linewidth=1.0,)               # draw and label parallels (latitude lines) for values (in degrees) given in the sequence circles
map1.drawmeridians(np.arange(-180,180,20),color='k',linewidth=1.0,)             # draw and label meridians (longitude lines) for values (in degrees) given in the sequence meridians
x,y=map1(longitude,latitude,inverse=False)                                      # get x and y coordinates for given longitude and latitude
plt.plot(x,y,'r*')                                                              # make plot center dot
plt.title('Azimuthal Equidistant Projection')                                   # add title to plot
plt.show()                                                                      # make plot show
```

The relevant Basemap methods are:

* **drawcoastlines()**: draw coastlines.
* **fillcontinents()**: color the interior of continents (by filling the coastline polygons). Unfortunately, the fillcontinents method doesn't always do the right thing. Matplotlib always tries to fill the inside of a polygon. Under certain situations, what is the inside of a coastline polygon can be ambiguous, and the outside may be filled instead of the inside. In these situations, the recommended workaround is to use the
* **drawlsmask()**: method to overlay an image with different colors specified for land and water regions.
* **drawcountries()**: draw country boundaries.
* **drawstates()**: draw state boundaries in North America.
* **drawrivers()**: draw rivers.

Source: https://matplotlib.org/basemap/users/geography.html

```python
# draw map with latitude and longitude
maps=Basemap(width=5000000,height=5000000,projection='lcc',lat_0=28.612894,lon_0=77.229446,resolution='l')# load map for given values
plt.figure(figsize=(10,8))                                                      # adjust the size if needed
maps.drawcoastlines()                                                           # draw coastlines.
maps.drawcountries()                                                            # make draw countries
maps.drawstates()                                                               # make draw states (NOT WORKING)
maps.drawrivers(linewidth=0.5,linestyle='solid',color='blue')                   # draw rivers (TAKES LOT OF TIME)
# draw a boundary around the map, fill the background. This background will end up being the ocean color, since the continents
#   will be drawn on top.
maps.drawmapboundary(fill_color='aqua')                                         # draw a boundary around the map
maps.fillcontinents(color='#70543e',lake_color='skyblue')                       # fill continents, set lake color same as ocean color
maps.drawparallels(np.arange(0,45,4))                                           # draw and label parallels (latitude lines)
maps.drawmeridians(np.arange(60,110,4))                                         # draw and label meridians (longitude lines)
plt.show()                                                                      # make show map
```

```python
# make draw map using -> drawlsmask() method to overlay an image with different colors specified for land and water regions.
maps=Basemap(width=5000000,height=5000000,projection='lcc',lat_0=28.612894,lon_0=77.229446,resolution='l')# load map for given values
maps.drawlsmask(land_color='lawngreen',ocean_color='skyblue',lakes=True,resolution='i')# make draw mask
plt.show()                                                                      # make show map
```

Instead of drawing coastlines and political boundaries, an image can be used as a map background. Basemap provides several options for this:

* **drawlsmask()**: draw a high-resolution land-sea mask as an image, with land and ocean colors specified. The land-sea mask is derived from the GSHHS coastline data, and there are several coastline options and pixel sizes to choose from.
* **bluemarble()**: draw a NASA Blue Marble image as a map background.
* **shadedrelief()**: draw a shaded relief image as a map background.
* **etopo()**: draw an etopo relief image as map background.
* **warpimage()**: use an abitrary image as a map background. The image must be global, covering the world in lat/lon coordinates from the international dateline eastward and the South Pole northward.

```python
# make draw map using -> bluemarble(): draw a NASA Blue Marble image as a map background.
maps=Basemap(width=5000000,height=5000000,projection='lcc',lat_0=28.612894,lon_0=77.229446,resolution='l')# load map for given values
maps.bluemarble()                                                               # set backgroud type
maps.drawcountries()                                                            # make draw countries
plt.show()                                                                      # make show map
```

```python
# make draw map using -> shadedrelief(): draw a shaded relief image as a map background
maps=Basemap(width=5000000,height=5000000,projection='lcc',lat_0=28.612894,lon_0=77.229446,resolution='l')# load map for given values
maps.shadedrelief()                                                             # set backgroud type
plt.show()                                                                      # make show map
```

```python
# make draw map using -> etopo(): draw an etopo relief image as map background.
maps=Basemap(width=5000000,height=5000000,projection='lcc',lat_0=28.612894,lon_0=77.229446,resolution='l')# load map for given values
maps.etopo()                                                                    # set backgroud type
plt.show()                                                                      # make show map
```

```python
# make draw map using -> warpimage(): use an abitrary image as a map background. The image must be global, covering the world in
#   lat/lon coordinates from the international dateline eastward and the South Pole northward.
maps=Basemap(resolution='i')                                                    # load map (world)
plt.figure(figsize=(15,8))                                                      # adjust the size if needed
maps.warpimage()                                                                # set backgroud type
maps.drawcoastlines(linewidth=0.5,color='black',linestyle='-.')                 # draw boundaries
plt.show()                                                                      # make show plot
```
