# [Probability Distribution](https://en.wikipedia.org/wiki/Probability_distribution)

A probability distribution is the mathematical function that gives the probabilities of occurrence of different possible outcomes for an experimen

## [Normal Distribution](https://en.wikipedia.org/wiki/Normal_distribution) (or Gaussian Distribution)

![png](https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Standard_deviation_diagram_micro.svg/1280px-Standard_deviation_diagram_micro.svg.png)

**Image: Continuous probability distribution**



image.png

The parameter &mu;  is the mean or expectation of the distribution (and also its median and mode), while the parameter &sigma; is its standard deviation. The variance of the distribution is &sigma;<sup>2</sup>. A random variable with a Gaussian distribution is said to be normally distributed, and is called a normal deviate.

### generate 1000 random values from standard normal probability distribution using - [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html)

![svg](https://upload.wikimedia.org/wikipedia/commons/7/74/Normal_Distribution_PDF.svg)

<br>
A selection of Normal Distribution Probability Density Functions (PDFs). Both the mean, μ, and variance, σ², are varied. The key is given on the graph.

**NOTE: The red curve is the standard normal distribution. For standard normal distribution mean (μ) is 0.0 and variance (σ²) is 1.0.**

```python
# generate 1000 random values with mean -> 0.0 and standard deviation -> 1.0 (i.e., from standard normal probability distribution)
standard_normal_distribution=np.random.normal(loc=0.0,scale=1.0,size=1000)      # return numpy array
# source: https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
# Parameters
#   loc : float or array_like of floats Mean ("centre") of the distribution.
#   scale : float or array_like of floats Standard deviation (spread or "width") of the distribution.
```

```python
# see information of `standard_normal_distribution`
print('Sample array dtype:',standard_normal_distribution.dtype)                 # data type in numpy array
print('Sample array shape:',standard_normal_distribution.shape)                 # shape of array (rows, columns)
print('Sample array number of dimension:',standard_normal_distribution.ndim)    # number of dimensions
print('Maximum value in array:',standard_normal_distribution.max())             # maximum value in array
print('Minimum value in array:',standard_normal_distribution.min())             # minimum value in array
print('Mean value of array:',standard_normal_distribution.mean())               # mean value of array
print('Standard deviation value of array:',standard_normal_distribution.std())  # standard deviation of the array elements
```

```python
# make data plot using histogram, source: https://seaborn.pydata.org/generated/seaborn.histplot.html
# A histogram is an approximate representation of the distribution of numerical data. Source: https://en.wikipedia.org/wiki/Histogram
sns.histplot(x=standard_normal_distribution,                                    # make plot data at x-axis
             color='red',                                                       # set color of plot
             kde=True,                                                          # make plot kernel density estimation (kde)
             stat='probability',                                                # statistics of plot
             bins='auto',                                                       # bins numbers or size
             element='bars'                                                     # type of plot
             )
# Kernel density estimation (KDE) is the application of kernel smoothing for probability density
#  estimation, i.e., a non-parametric method to estimate the probability density function of a random
#  variable based on kernels as weights. Source: https://en.wikipedia.org/wiki/Kernel_density_estimation
plt.grid()                                                                      # make plot grid
plt.xlabel('Data Values (Random Values)')                                       # add x-axis label
plt.title('Standard Normal Distribution')                                       # add title
plt.show()                                                                      # make show plot
# element : {"bars", "step", "poly"}
#   Visual representation of the histogram statistic. Only relevant with univariate data.
# fill : bool If True, fill in the space under the histogram.
# kde : bool, If True, compute a kernel density estimate to smooth the distribution
#    and show on the plot as (one or more) line(s).
# stat : str
#    Aggregate statistic to compute in each bin.
#       count: (default) show the number of observations in each bin
#       frequency: show the number of observations divided by the bin width
#       probability or proportion: normalize such that bar heights sum to 1
#       percent: normalize such that bar heights sum to 100
#       density: normalize such that the total area of the histogram equals 1
# bins : str, number, vector, or a pair of such values
```

### generate 1 million random from standard normal probability distribution values using - [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html)

```python
# generate 1,000,000 (1 million) random values (big array) with mean -> 0.0 and standard deviation -> 1.0 (i.e., standard normal probability distribution)
standard_normal_distribution_1m=np.random.normal(loc=0.0,scale=1.0,size=1_000_000)# return numpy array
# see information of `standard_normal_distribution_1m`
print('Sample array dtype:',standard_normal_distribution_1m.dtype)              # data type in numpy array
print('Sample array shape:',standard_normal_distribution_1m.shape)              # shape of array (rows, columns)
print('Sample array number of dimension:',standard_normal_distribution_1m.ndim) # number of dimensions
print('Maximum value in array:',standard_normal_distribution_1m.max())          # maximum value in array
print('Minimum value in array:',standard_normal_distribution_1m.min())          # minimum value in array
print('Mean value of array:',standard_normal_distribution_1m.mean())            # mean value of array
print('Standard deviation value of array:',standard_normal_distribution_1m.std())# standard deviation of the array elements
```

```python
# make data plot using histogram
sns.histplot(x=standard_normal_distribution_1m,                                 # make plot data at x-axis
             color='green',                                                     # set color of plot
             kde=False,                                                         # make plot kernel density estimation (kde)
             stat='probability',                                                # statistics of plot
             bins='auto',                                                       # bins numbers or size
             element='poly',                                                    # type of plot
             alpha=0.1                                                          # adjust transparency , 1.0 -> No transparency
             )
plt.grid()                                                                      # make plot grid
plt.xlabel('Data Values (Random Values)')                                       # add x-axis label
plt.title('Standard Normal Distribution (for 1 million random values)')         # add title
plt.show()                                                                      # make show plot
```

### generate random values from multiple normal probability distribution using - [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) like -

![svg](https://upload.wikimedia.org/wikipedia/commons/7/74/Normal_Distribution_PDF.svg)

```python
# generate 1,000,000 (1 million) random values with different mean and standard deviation
normal_distribution_l0s0_2=np.random.normal(loc=0.0,scale=np.sqrt(0.2),size=1_000_000)# mean -> 0.0 and standard deviation -> √0.2
normal_distribution_l0s1=np.random.normal(loc=0.0,scale=np.sqrt(1.0),size=1_000_000)# mean -> 0.0 and standard deviation -> √1.0
normal_distribution_l0s5=np.random.normal(loc=0.0,scale=np.sqrt(5.0),size=1_000_000)# mean -> 0.0 and standard deviation -> √5.0
normal_distribution_l__2s0_5=np.random.normal(loc=-2.0,scale=np.sqrt(0.5),size=1_000_000)# mean -> -2.0 and standard deviation -> √0.5
```

```python
# make plot all generated normal probability distribution plot using histogram
plt.figure(figsize=(15,7))                                                      # make plot and set figure size
# source (matplotlib.pyplot.figure): https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
# figsize(float, float), default: rcParams["figure.figsize"] (default: [6.4, 4.8]) - width (x-axis) and height (y-axis) in inches.
# make plot normal distribution for mean -> 0.0 and standard deviation -> √0.2
sns.histplot(x=normal_distribution_l0s0_2,color='blue',kde=False,stat='density',bins='auto',element='poly',alpha=0.1)
plt.axvline(x=normal_distribution_l0s0_2.mean(),label='mean = 0.0 , standard deviation = √0.2',color='b',ls='--',lw=0.5)# make vertical line to mark mean
# here (matplotlib.pyplot.axvline) (source: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axvline.html):
#   ls or linestyle : line style ; supported values are : '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', and 'dotted'
#   lw or linewidth : line width

# make plot normal distribution formean -> 0.0 and standard deviation -> √1.0
sns.histplot(x=normal_distribution_l0s1,color='red',kde=False,stat='density',bins='auto',element='poly',alpha=0.1)
plt.axvline(x=normal_distribution_l0s1.mean(),label='mean = 0.0 , standard deviation = √1.0',color='r',ls='--',lw=0.5)# make vertical line to mark mean

# make plot normal distribution for mean -> 0.0 and standard deviation -> √5.0
sns.histplot(x=normal_distribution_l0s5,color='yellow',kde=False,stat='density',bins='auto',element='poly',alpha=0.1)
plt.axvline(x=normal_distribution_l0s5.mean(),label='mean = 0.0 , standard deviation = √5.0',color='yellow',ls='--',lw=0.5)# make vertical line to mark mean

# make plot normal distribution for mean -> -2.0 and standard deviation -> √0.5
sns.histplot(x=normal_distribution_l__2s0_5,color='green',kde=False,stat='density',bins='auto',element='poly',alpha=0.1)
plt.axvline(x=normal_distribution_l__2s0_5.mean(),label='mean = -2.0 , standard deviation = √0.5',color='g',ls='--',lw=0.5)# make vertical line to mark mean

plt.legend()                                                                    # make plot labels space
plt.grid()                                                                      # make plot grid
plt.xlabel('Data Values (Random Values)')                                       # add x-axis label
plt.title('Normal Probability Distributions (for 1 million random values)')     # add title
plt.savefig('normal_probability_distributions.png',dpi=480)                     # make save figure
plt.show()                                                                      # make show plot
```

### generate 1000 random values from continuous uniform probability distribution using - [numpy.random.uniform](https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html)

```python
# generate 1000 random values from continuous uniform probability distribution with lower limit = 0.0 and higher = 1.0
continuous_uniform_distribution=np.random.uniform(low=0.0,high=1.0,size=1000)   # return numpy array
# source: https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html
# Parameters
#   low : float or array_like of floats, optional Lower boundary of the output interval.  All values generated will be
#     greater than or equal to low. The default value is 0.
#   high : float or array_like of floats Upper boundary of the output interval.  All values generated will be less than or equal to high.
#     The high limit may be included in the returned array of floats due to floating-point rounding in the  equation
#     ``low + (high-low) * random_sample()``.  The default value
```

```python
# see information of `continuous_uniform_distribution`
print('Sample array dtype:',continuous_uniform_distribution.dtype)              # data type in numpy array
print('Sample array shape:',continuous_uniform_distribution.shape)              # shape of array (rows, columns)
print('Sample array number of dimension:',continuous_uniform_distribution.ndim) # number of dimensions
print('Maximum value in array:',continuous_uniform_distribution.max())          # maximum value in array
print('Minimum value in array:',continuous_uniform_distribution.min())          # minimum value in array
print('Mean value of array:',continuous_uniform_distribution.mean())            # mean value of array
print('Standard deviation value of array:',continuous_uniform_distribution.std())# standard deviation of the array elements
```

```python
# make plot uniform probability distribution - for a -> 0.0 and b -> 1.0 (1000 samples)
sns.histplot(x=continuous_uniform_distribution,color='red',kde=True,stat='count',bins='auto',element='bars')# make plot distribution
plt.axvline(x=continuous_uniform_distribution.mean(),label='mean 1000',color='r',ls='--',lw=1.0)# make plot mean
plt.grid()                                                                      # make plot grid
plt.xlabel('Data Values (Random Values)')                                       # add x-axis label
plt.title('Continuous Uniform Probability Distribution')                        # add title
plt.show()                                                                      # make show plot
```

### generate 1 million random values from continuous uniform probability distribution using - [numpy.random.uniform](https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html)

```python
# generate 1,000,000 (1 million) random values from continuous uniform probability distribution with lower limit = 0.0 and higher = 1.0
continuous_uniform_distribution_1m=np.random.uniform(low=0.0,high=1.0,size=1_000_000)# return numpy array
# see information of `continuous_uniform_distribution_1m`
print('Sample array dtype:',continuous_uniform_distribution_1m.dtype)           # data type in numpy array
print('Sample array shape:',continuous_uniform_distribution_1m.shape)           # shape of array (rows, columns)
print('Sample array number of dimension:',continuous_uniform_distribution_1m.ndim)# number of dimensions
print('Maximum value in array:',continuous_uniform_distribution_1m.max())       # maximum value in array
print('Minimum value in array:',continuous_uniform_distribution_1m.min())       # minimum value in array
print('Mean value of array:',continuous_uniform_distribution_1m.mean())         # mean value of array
print('Standard deviation value of array:',continuous_uniform_distribution_1m.std())# standard deviation of the array elements
```

```python
# make plot uniform probability distribution - for a -> 0.0 and b -> 1.0 for 1,000,000 (1 million)
sns.histplot(x=continuous_uniform_distribution_1m,color='red',kde=True,stat='count',bins='auto',element='bars')# make plot distribution
plt.axvline(x=continuous_uniform_distribution_1m.mean(),label='mean 1000',color='r',ls='--',lw=1.0)# make plot mean
plt.grid()                                                                      # make plot grid
plt.xlabel('Data Values (Random Values)')                                       # add x-axis label
plt.title('Continuous Uniform Probability Distribution (for 1 million random values)')# add title
plt.show()                                                                      # make show plot
```

### generate random values from multiple continuous uniform probability distribution using - [numpy.random.uniform](https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html)

```python
# generate 1,000,000 (1 million) random values with different lower and higher values
continuous_uniform_distribution_l__1h1=np.random.uniform(low=-1.0,high=1.0,size=1_000_000)# lower = -1.0 and higher = 1.0
continuous_uniform_distribution_l__0_5h0_5=np.random.uniform(low=-0.5,high=0.5,size=1_000_000)# lower = -0.5 and higher = 0.5
continuous_uniform_distribution_l2h5=np.random.uniform(low=2.0,high=5.0,size=1_000_000)# lower = 2.0 and higher = 5.0
continuous_uniform_distribution_l__4h__2=np.random.uniform(low=-4.0,high=-2.0,size=1_000_000)# lower = -4.0 and higher = -2.0
```

```python
# make plot all uniform probability distribution plot using histogram
plt.figure(figsize=(15,7))                                                      # make plot and set figure size
# make plot uniform distribution for a (lower) -> -1.0 and b (higher) -> 1.0
sns.histplot(x=continuous_uniform_distribution_l__1h1,color='red',kde=False,stat='density',bins='auto',element='poly',alpha=0.1)
plt.axvline(x=continuous_uniform_distribution_l__1h1.mean(),label='a (lower) -> -1.0 and b (higher) -> 1.0',color='r',ls='--',lw=0.5)# make vertical line to mark mean

# make plot uniform distribution for a (lower) -> -0.5 and b (higher) -> 0.5
sns.histplot(x=continuous_uniform_distribution_l__0_5h0_5,color='blue',kde=False,stat='density',bins='auto',element='poly',alpha=0.1)
plt.axvline(x=continuous_uniform_distribution_l__0_5h0_5.mean(),label='a (lower) -> -0.5 and b (higher) -> 0.5',color='b',ls='--',lw=0.5)# make vertical line to mark mean

# make plot uniform distribution for a (lower) -> 2.0 and b (higher) -> 5.0
sns.histplot(x=continuous_uniform_distribution_l2h5,color='green',kde=False,stat='density',bins='auto',element='poly',alpha=0.1)
plt.axvline(x=continuous_uniform_distribution_l2h5.mean(),label='a (lower) -> 2.0 and b (higher) -> 5.0',color='g',ls='--',lw=0.5)# make vertical line to mark mean

# make plot uniform distribution for a (lower) -> -4.0 and b (higher) -> -2.0
sns.histplot(x=continuous_uniform_distribution_l__4h__2,color='yellow',kde=False,stat='density',bins='auto',element='poly',alpha=0.1)
plt.axvline(x=continuous_uniform_distribution_l__4h__2.mean(),label='a (lower) -> -4.0 and b (higher) -> -2.0',color='yellow',ls='--',lw=0.5)# make vertical line to mark mean

plt.legend()                                                                    # make plot labels space
plt.grid()                                                                      # make plot grid
plt.xlabel('Data Values (Random Values)')                                       # add x-axis label
plt.title('Uniform Probability Distributions (for 1 million random values)')    # add title
plt.savefig('uniform_probability_distributions.png',dpi=480)                    # make save figure
plt.show()                                                                      # make show plot
```
