# Plotting with [Matplotlib](https://matplotlib.org/) and [Seaborn](https://seaborn.pydata.org/)

### simple plots

#### linear line plots

Official Docs: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# make points - A(-1,5), B(1,2), C(3,4), D(4,9), E(10,11)
x=[-1,1,3,4,10]                                                                 # x-axis
y=[5,2,4,9,11]                                                                  # y-axis
```

**simple plot**

```python
# colors: https://matplotlib.org/stable/gallery/color/named_colors.html
# or html color codes - like "#000011" # or # get code - https://htmlcolorcodes.com/
# markers - https://matplotlib.org/stable/api/markers_api.html
plt.plot(x,y,color='red',marker='*',label='line')
plt.legend()                                                                    # make spaces for labels
plt.grid()                                                                      # for grid lines
plt.xlabel('x-axis')                                                            # add label to x-axis
plt.ylabel('y-axis')                                                            # add label to y-axis
plt.title('Simple Plot')                                                        # add title to plot
plt.savefig('sample figure.png',dpi=480)                                        # make figure save
plt.show()                                                                      # show plot
```

**multiple simple plot**

```python
# colors: https://matplotlib.org/stable/gallery/color/named_colors.html
# or html color codes - like "#000011" # or # get code - https://htmlcolorcodes.com/
# markers - https://matplotlib.org/stable/api/markers_api.html
plt.plot(x,y,color='red',marker='*',label='line',linewidth=2)# plot 1 in same figure
plt.plot(np.square(x),y,color='green',marker='1',label='square of line')# plot 2 in same figure
plt.legend()                                                                    # make spaces for labels
plt.grid()                                                                      # for grid lines
plt.xlabel('x-axis')                                                            # add label to x-axis
plt.ylabel('y-axis')                                                            # add label to y-axis
plt.title('Simple Plot')                                                        # add title to plot
plt.savefig('sample figure.png',dpi=480)                                        # make figure save
plt.show()                                                                      # show plot
```

```python
plt.plot(x,y,color='red',marker='*',label='line')# plot 1 in same figure
plt.legend()                                                                    # make spaces for labels
plt.grid()                                                                      # for grid lines
plt.show()                                                                      # show plot
# plot one ends here

plt.plot(np.square(x),y,color='green',marker='1',label='square of line')        # plot 2 in same figure
plt.legend()                                                                    # make spaces for labels
plt.xlabel('x-axis')                                                            # add label to x-axis
plt.ylabel('y-axis')                                                            # add label to y-axis
plt.title('Simple Plot')                                                        # add title to plot
plt.show()                                                                      # show plot
# plot 2 ends here
```

#### non-linear line plots and scatter plots

Official Docs (scatter): https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
# make plot sine plot - x (angles)
x=np.linspace(-360,360,100,endpoint=True)# generate 100 points between -360 and 360
y=np.sin(np.deg2rad(x))                                                         # find sine of x (all points)
```

**simple plot**

```python
plt.plot(x,y,color='red',label='sine',linewidth=5,linestyle='--')               # plot
plt.legend()                                                                    # make spaces for labels
plt.grid()                                                                      # for grid lines
plt.xlabel('degree')                                                            # add label to x-axis
plt.ylabel('sine')                                                              # add label to y-axis
plt.title('Sine Plot')                                                          # add title to plot
plt.show()                                                                      # show plot
```

**multiple plots**

```python
plt.plot(x,y,color='red',label='sine',lw=1,ls='-.')                             # plot 1 in same figure
plt.scatter(x,np.cos(np.deg2rad(x)),color='b',label='cosine',s=1)               # plot 2 in same figure
plt.legend()                                                                    # make spaces for labels
plt.grid()                                                                      # for grid lines
plt.xlabel('degree')                                                            # add label to x-axis
plt.ylabel('value')                                                             # add label to y-axis
plt.title('Sine and Cosine Plot')                                               # add title to plot
plt.show()                                                                      # show plot
```

```python
# make subplot (divide plot) - parts - 2 row, 1 column
plt.subplot(2,1,2)                                                              # 2 rows, 1 column, plot number - 2
plt.plot(x,y,color='red',label='sine',lw=1,ls='-.')# plot 1 in same figure
plt.axis('off')                                                                 # make axis off
plt.legend()                                                                    # make spaces for labels
# now plot 1 ends when we again call subplot
plt.subplot(2,1,1)                                                              # 2 rows, 1 column, plot number - 1
plt.scatter(x,np.cos(np.deg2rad(x)),color='b',label='cosine',s=1)               # plot 2 in same figure
plt.grid()                                                                      # for grid lines
plt.xlabel('degree')                                                            # add label to x-axis
plt.ylabel('value')                                                             # add label to y-axis
plt.title('Cosine Plot')                                                        # add title to plot
# make plot show (All ends)
plt.show()                                                                      # show plot
```

#### make random circles with scatter plot

```python
# generate points with uniform distribution
# Source (uniform distribution): https://en.wikipedia.org/wiki/Continuous_uniform_distribution
points=np.random.uniform(low=-1,high=1,size=200).reshape(-1,2)# reshape to 2D
# random sizes of circles
sizes=np.linspace(50,500,100)                                                   # generate random sizes between 50 and 500 - 100 points
#np.random.shuffle(sizes)                                                       # shuffle sizes (shuffle elements in array)
# NOTE: shuffle orginal array (randomly changes order of elements)
# make circles (scatter plot)
plt.scatter(points[:,0],points[:,1],c=np.random.uniform(low=0,high=1,size=100),
            s=sizes,alpha=0.4,cmap=plt.cm.viridis)
# alpha - transparent factor (between 0 (full transparent) and 1 (full color))
plt.show()                                                                      # make show
```

### plot probability distribution - [Uniform](https://en.wikipedia.org/wiki/Continuous_uniform_distribution) and [Standard (Gaussian)](https://en.wikipedia.org/wiki/Standard_deviation)  

**Uniform Distribution**

by `numpy.random.uniform`

![Uniform Distribution](https://upload.wikimedia.org/wikipedia/commons/thumb/9/96/Uniform_Distribution_PDF_SVG.svg/480px-Uniform_Distribution_PDF_SVG.svg.png?20130723141236)

```python
# generate data
uniform_points=np.random.uniform(low=-1,high=1,size=10_000)
# make plot - histogram
# Source: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
plt.hist(x=uniform_points,bins=None,histtype='bar',orientation='vertical',
         color='red',alpha=0.8)
plt.grid()                                                                      # add grid lines
plt.show()                                                                      # show plot
```

**Standard Probability Distribution**

by `numpy.random.normal`

![Standard deviation diagram](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/Standard_deviation_diagram.svg/400px-Standard_deviation_diagram.svg.png)

```python
# generate data # loc -> mean, scale -> standard deviation
normal_points=np.random.normal(loc=0,scale=1,size=10_000)
# make plot - histogram
# Source: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html
plt.hist(x=normal_points,bins=None,histtype='bar',orientation='vertical',
         color='red',alpha=0.6)
plt.grid()                                                                      # add grid lines
plt.show()                                                                      # show plot
```

#### [count plot](https://seaborn.pydata.org/generated/seaborn.countplot.html)

```python
# count plot - count values (unique) in column (categorical values)
sns.countplot(x=tips_df['smoker'])
plt.grid()                                                                      # add grid
plt.show()                                                                      # make plot show
```

```python
# count plot - count values (unique) in column (categorical values)
#sns.countplot(x=tips_df['smoker'],hue=tips_df['sex'])
sns.countplot(data=tips_df,x='smoker',hue='sex')
plt.grid()                                                                      # add grid
plt.show()                                                                      # make plot show
```

#### [histogram](https://seaborn.pydata.org/generated/seaborn.histplot.html)

```python
# histogram plot - https://seaborn.pydata.org/generated/seaborn.histplot.html
sns.histplot(data=tips_df,y='tip',stat='density',color='red',kde=True)
# kde (kernel density estimation) - https://en.wikipedia.org/wiki/Kernel_density_estimation
# Stats - Aggregate statistic to compute in each bin.
#     count: show the number of observations in each bin
#     frequency: show the number of observations divided by the bin width
#     probability or proportion: normalize such that bar heights sum to 1
#     percent: normalize such that bar heights sum to 100
#     density: normalize such that the total area of the histogram equals 1
plt.grid()                                                                      # add grid
plt.show()                                                                      # make plot show
```

```python
# histogram plot
sns.histplot(data=tips_df,x='day',y='tip')# as continues
plt.grid()                                                                      # add grid
plt.show()                                                                      # make plot show
```
